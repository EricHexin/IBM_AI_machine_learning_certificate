```python
import pandas as pd
import numpy as np


from sklearn.metrics import mean_squared_error
```

```python
concrete_data = pd.read_csv('concrete_data.csv')
concrete_data.head()
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1] # number of predictors
n_cols
```

# A: Build a baseline model (5 marks) 

```python
from sklearn.model_selection import train_test_split
predictors_train, predictors_test, target_train, target_test = train_test_split( predictors, target, test_size=0.3, random_state=42)
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = regression_model()

# fit the model
model.fit(predictors_train, target_train, validation_data=(predictors_test, target_test), epochs=50, verbose=2)
```

```python

scores = model.evaluate(predictors_test, target_test, verbose=0)
print("Accuracy: {} \n Error: {}".format(scores, 100-scores*100))
```

```python
prd = model.predict(predictors_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(target_test, prd)
```

```python
MSEs = []

for nn in range(50):
    predictors_train, predictors_test, target_train, target_test = train_test_split( predictors, target, test_size=0.3, random_state=42)

    model.fit(predictors_train, target_train, validation_data=(predictors_test, target_test), epochs=50, verbose=2)

    prd = model.predict(predictors_test)

    MSEs += [mean_squared_error(target_test, prd)]
```

```python
# mean and standard deviation of mean squared errors
np.mean(MSEs), np.std(MSEs)
```

# D: repeat with 3 hidden layers

```python
# define regression model
def regression_model3():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model3 = regression_model3()
```

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
# predictors_norm.head()
```

```python
for nn in range(50):
    predictors_train, predictors_test, target_train, target_test = train_test_split( predictors_norm, target, test_size=0.3, random_state=42)

    model3.fit(predictors_train, target_train, validation_data=(predictors_test, target_test), epochs=50, verbose=2)

    prd = model3.predict(predictors_test)

    MSEs += [mean_squared_error(target_test, prd)]
```

```python
np.mean(MSEs), np.std(MSEs)
```
