```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
n_cols = predictors.shape[1] # number of predictors
```

```python
n_cols
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.30, random_state=44)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
keras.backend.clear_session() #Destroys the current TF graph and creates a new one. Useful to avoid clutter from old modelslayers.
```

```python
# build the model
model = regression_model()
```

```python
from matplotlib import pyplot
```

```python
from sklearn.metrics import mean_squared_error
```

```python
lst_mean_sq_erro=[]
for i in range(49):
    keras.backend.clear_session() #Destroys the current TF graph and creates a new one. Useful to avoid clutter from old modelslayers.
    model = regression_model()
    model.fit(X_train, y_train,  epochs=50)
    y_pred=model.predict(X_test)
    lst_mean_sq_erro.append(mean_squared_error(y_test,y_pred))
```
